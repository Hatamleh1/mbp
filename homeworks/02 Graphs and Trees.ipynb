# Task 1

For a given graph find an Eulerian path and write it as a sequence of nodes.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from("ABCDEFG")
all_edges = ["AB", "BC", "CD", "DE", "EF", "FG", "GA", "AC", "CE", "AF", "FC", "EB", "BG"]
G.add_edges_from(all_edges)
nx.draw(G, with_labels = True)
plt.show()
```

```python
your_answer = "ABCDEFGM"

assert len(your_answer) -1 == len(all_edges), f"Number of edges is {len(all_edges)}, not {len(your_answer) - 1}"

for i in range(len(your_answer) - 1):
    f = your_answer[i:i+2]
    r = f[::-1]
    print(r)
    assert f in all_edges or r in all_edges, f"No such edge {r}"
```
